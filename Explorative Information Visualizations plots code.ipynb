```python
import math
import random
import numpy as np
import pandas as pd
import statistics
import scipy.stats as stats
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
import csv
```

```python
# Plot 1 - Pie Chart "Plastic Production by Sectors"
sectors = []
amount = []

with open('plastic-production-by-sector-cleaned_1.csv', 'r') as csvfile:
    lines = csv.reader(csvfile, delimiter = ',')
    for row in lines:
        sectors.append(row[0])
        amount.append(float(row[1]))
        

figure(figsize=(10, 10))
plt.pie(amount,labels = sectors,autopct = '%.2f%%')
plt.title('Plastic Production by Sectors', fontsize = 17)
plt.show()
```

```python
# Plot 2 - "Global Plastic Production from the year 1950-2019"
x = []
y = []
  
with open('global-plastic.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        x.append(row[0])
        y.append(int(row[1]))

figure(figsize=(12, 5))
plt.plot(x, y, color = 'b', linestyle = 'dashed',
         marker = 'o',label = "plastic production")

plt.xticks(np.arange(0, 70, 4))
plt.xlabel('year')
plt.ylabel('plastic production')
plt.title('Global Plastic Production from the year 1950-2019', fontsize = 15)
plt.legend()
plt.show()
```

```python
# Plot 3 - Pie Chart "Plastic Production by Types"
sectors = []
amount = []

with open('plastic-production-by-sector-cleaned_2.csv', 'r') as csvfile:
    lines = csv.reader(csvfile, delimiter = ',')
    for row in lines:
        sectors.append(row[0])
        amount.append(float(row[1]))
        

figure(figsize=(10, 10))
plt.pie(amount,labels = sectors,autopct = '%.2f%%')
plt.title('Plastic Production by Types', fontsize = 17)
plt.show()
```

```python
# Plot 4 - "Mismanaged Plastic Waste in 2019 top 10 Countries"
df = pd.read_csv('plastic-pollution.csv')

sns.set(rc={'figure.figsize':(16,9)})

g = df.groupby('Country', as_index=False)['Mismanaged plastic waste to ocean per capita (kg per year)'].sum().sort_values(by='Mismanaged plastic waste to ocean per capita (kg per year)', ascending=False).head(10)
sns.barplot(data=g, x='Country', y='Mismanaged plastic waste to ocean per capita (kg per year)', hue='Country', dodge=False).set(xticklabels=[]);
```
